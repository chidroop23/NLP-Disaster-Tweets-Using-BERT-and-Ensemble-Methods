```python
import pandas as pd
import re
import string
import numpy as np
```

```python
train = pd.read_csv('train.csv', index_col=0)

test = pd.read_csv('test.csv', index_col=0)
```

```python
# fill in missign value 
for df in [train, test]:
    for col in ['keyword', 'location']:
        df[col] = df[col].fillna(f'no_{col}')
```

```python
# remove url and username inside comments

def remove_url(text):
    text = re.sub('\\b((?:https?|ftp|file)://[-a-zA-Z0-9+&@#/%?=~_|!:, .;]*[-a-zA-Z0-9+&@#/%=~_|])','', text)
    return text

train['text'] = train['text'].apply(remove_url)
test['text'] = test['text'].apply(remove_url)
```

```python
# remove html format

def remove_html(text):
    text = re.sub(r'<.*?>', '', text)
    return text

train['text'] = train['text'].apply(remove_html)
test['text'] = test['text'].apply(remove_html)
```

```python
# remove references within tweets
def remove_references(text):
    text = re.sub(r'&[a-zA-Z]+;?', '', text)
    return text

train['text'] = train['text'].apply(remove_references)
test['text'] = test['text'].apply(remove_references)
```

```python
# remove numbers
def remove_num(text):
    text = re.sub(r'\w*\d+\w*', '', text)
    return text

train['text'] = train['text'].apply(remove_num)
test['text'] = test['text'].apply(remove_num)
```

```python
# set tweet content to lower case

train['text'] = [t.lower() for t in train['text']]
test['text'] = [t.lower() for t in test['text']]
```

```python
# remove repeated component
def remove_repeated(text):
    elongated = re.compile(r'(\S*?)([a-z])\2{2,}')
    text = elongated.sub(r'\1\2', text)
    return text

train['text'] = train['text'].apply(remove_repeated)
test['text'] = test['text'].apply(remove_repeated)
```

```python
# remove mentioned users
def remove_mention(text):
    text = re.sub(r'@\w+', '', text)
    return text

train['text'] = train['text'].apply(remove_mention)
test['text'] = test['text'].apply(remove_mention)
```

```python
# remove punctuation
def remove_punct(text):
    text = re.sub(r'[^\w\s]', '', text)
    return text

train['text'] = train['text'].apply(remove_punct)
test['text'] = test['text'].apply(remove_punct)
```

```python
# remove non ascii printable
string.printable

def remove_non_printable(text):
    text = ''.join([word for word in text if word in string.printable])
    return text

train['text'] = train['text'].apply(remove_non_printable)
test['text'] = test['text'].apply(remove_non_printable)
```

```python
# remove numbers
def remove_num(text):
    text = re.sub(r'\w*\d+\w*', '', text)
    return text

train['text'] = train['text'].apply(remove_num)
test['text'] = test['text'].apply(remove_num)
```
